```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots 
```

```python
g=9.81 #m/s2
qin0 = 0.1 # m3/s
qin = 1.1*qin0 #m3/s

Ac = 1. #m2
Ahole = 0.01 #m2
ca_in0 = 1. #mol/m3
ca_in = 1.25*ca_in0 #mol/m3

ca0=ca_in0
h0 = ((qin0/Ahole)**2)/2/g
tend = 2000. #s
```

```python
def dvdt(t,v):
    h,ca = v
    qout = Ahole * (2*g*h)**0.5
    dhdt = (qin - qout)/Ac
    dcadt =  ((ca_in*qin - ca*qout)/Ac - ca*dhdt)/h
    return [dhdt, dcadt]
```

```python
res = solve_ivp(dvdt, (0,tend), [h0, ca0], dense_output=True, method='Radau')
tplot = np.linspace(0,tend,1000)
h,ca = res.sol(tplot)
fig = make_subplots(rows=1,cols=2)
fig.add_scatter(x=tplot, y=h,mode='lines',name='height',row=1,col=1)
fig.add_scatter(x=tplot, y=ca, mode='lines', name='ca', row=1,col=2)
fig.update_layout(height=600,width=1000,template='plotly_dark')
```
