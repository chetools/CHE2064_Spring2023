```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
q=100. #L/min
cai = 1. #mol/L
Ti = 350. #K
V = 100. #L
rho = 1000. #g/L
C = 0.239 #J/(g K)
negHR = 5e4 #J/mol
ER = 8750. #K
k0 = 7.2e10 #1/min
UA = 5e4 #J/(min K)
ca0 = 0.5 #mol/L
T0 = 350 #K
Tc=310. #K
tend = 10. #min
```

```python
def rhs(t, v):
    ca, T = v
    kca = ca*k0 * np.exp(-ER/T)
    dca = q*(cai - ca)/V - kca
    dT =  q*(Ti - T)/V + negHR*kca/(rho*C) - UA*(T-Tc)/(rho * V * C)
    return [dca, dT]
```

```python
res=solve_ivp(rhs, (0, tend), (ca0, T0), method='Radau',dense_output=True)
```

```python
tplot = np.linspace(0,tend, 1000)
ca, T = res.sol(tplot)
fig = make_subplots(rows=1, cols=2)
fig.add_scatter(x=tplot, y=ca, mode='lines',name='ca',row=1,col=1)
fig.add_scatter(x=tplot, y=T, mode='lines',name='T', row=1, col=2)
fig.update_layout(width=800,height=600, template='plotly_dark')
```
