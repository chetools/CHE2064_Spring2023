```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
g=9.81 #m/s**2
qin0 = 0.1 #m3/s
qin = 1.5*qin0
Ac1, Ac2, Ac3 =1., 0.5, 2. #m2
Ad = np.array([0.01, 0.015, 0.02]) #m2
tend = 1000.
```

```python
h0=qin0**2/((Ad**2)*2*g)
```

```python
def dhdt(t, h):
    #qin = qin0 + 0.2*qin0*np.sin(t/10)
    h1,h2,h3 = h
    q1,q2,q3= Ad*((2*g*h)**0.5)   #flow draining from each tank
    return [(qin - q1)/Ac1, (q1-q2)/Ac2, (q2-q3)/Ac3]
```

```python
res=solve_ivp(dhdt, (0,tend), h0,dense_output=True)
```

```python
tplot = np.linspace(0,tend,500)
h1,h2,h3=res.sol(tplot)
```

```python
fig=make_subplots()
fig.add_scatter(x=tplot,y=h1,name='h1',mode='lines')
fig.add_scatter(x=tplot,y=h2,name='h2',mode='lines')
fig.add_scatter(x=tplot,y=h3,name='h3',mode='lines')
fig.update_layout(height=600,width=800,template='plotly_dark')
```
