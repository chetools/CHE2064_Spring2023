```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
```

```python
g=9.81 #m/s**2
qin0 = 0.1 #m3/s
#qin = 1.5*qin0
Ac1, Ac2, Ac3 =1., 0.5, 2. #m2
Ad = np.array([0.01, 0.01, 0.01]) #m2
tend = 3000.
kc = 1e-2  #controller gain
taui = 3e2 #reset time for integral controller
h3sp = 8. #m

h0=qin0**2/((Ad**2)*2*g)
eint0 = 0.
v0 = np.concatenate([h0, np.array([eint0])])

def dvdt(t,v):
    h=v[:3]
    eint=v[3]
    h1,h2,h3=h
    e = h3sp-h3
    qin = qin0 + kc*(e + eint/taui)
    q1,q2,q3= Ad*((2*g*h)**0.5)   #flow draining from each tank
    return [(qin - q1)/Ac1, (q1-q2)/Ac2, (q2-q3)/Ac3, e]

res=solve_ivp(dvdt, (0,tend), v0 ,dense_output=True, method='Radau')
tplot = np.linspace(0,tend,500)
h1,h2,h3, eint=res.sol(tplot)
fig=make_subplots()
fig.add_scatter(x=tplot,y=h1,name='h1',mode='lines')
fig.add_scatter(x=tplot,y=h2,name='h2',mode='lines')
fig.add_scatter(x=tplot,y=h3,name='h3',mode='lines')
fig.update_layout(height=600,width=800,template='plotly_dark')
```
